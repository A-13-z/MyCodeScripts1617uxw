```python
corr_dct = {}
for col1 in train_file.columns:
    for col2 in train_file.columns:
        #print(train_file[col2].dtype)
        if col1 != col2 and (col2, col1) not in corr_dct.keys() and train_file[col1].dtype != object and train_file[col2].dtype != object:
            corr_dct[(col1, col2)] = train_file[col1].corr(train_file[col2])

print(corr_dct)
```

```python
col_rel_names = set()

for (k1, k2), v in corr_dct.items():
    if v > 0.3 or v < -0.3:
        col_rel_names.add(k1)
        col_rel_names.add(k2)
        print((k1, k2), v)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train = pd.DataFrame({col: train_file[col] for col in col_rel_names if col != 'loan_status'})
y_train = pd.DataFrame({'loan_status': train_file['loan_status']})
```

```python
X_test = pd.DataFrame({col: test_file[col] for col in col_rel_names if col != 'loan_status'})
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Initialize the model
model = XGBClassifier()

# Train the model
model.fit(X_train1, y_train1)

# Make predictions
y_pred1 = model.predict(X_test1)

# Evaluate the model
accuracy = accuracy_score(y_test1, y_pred1)
```
