# Assignment 3 - Data Estimation

The goal is to try to estimate various parameters given the raw data. The first part of the assignment is to use curve fitting to estimate 4 unknowns - speed of light (c), Planck's constant (h), boltzmann's constant (kb) and temperature (T). Note: In the estimations, boltzmann constant and temperature would be taken as a single unknown parameter.

## Estimating the parameters using curve_fit

Note:- The way the temperature is estimated is through dividing the estimation of the product of temperature and boltzman constant by the actual value of botzman constant. This is because there is no way to separately predict them, i.e. they are not independent of each other in this estimation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
```

```python
#function definition
def planck(t, c, h, kbt):
    t_ = [math.pow(i, 5) for i in t]
    return (2*h*c*c)/(t_*(np.exp(h*c/(np.multiply(kbt, t))) - 1))
```

```python
fle = open("d1.txt")    #opening the file
lnes = fle.readlines()  #reading the lines
lbda = [float(i.split(',')[0]) for i in lnes]   #extracting the x-axis values
b_val = [float(i.split(',')[-1][:-1]) for i in lnes]    #extracting the spectral rad values

#using curve fit
popt, pcov = curve_fit(planck, lbda, b_val, p0 = [3*math.pow(10, 8), 6.6*math.pow(10, -34), 4*math.pow(10, -20)])
#printing the optimum parameters
print("Speed of light: ", popt[0])
print("Planck's constant: ", popt[1])
print("Boltzman constant: ", 1.38e-23)
print("Temperature: ", popt[2]/1.38e-23)
#plotting the graph
plt.xlabel("wavelength")
plt.ylabel("Spectral radiance per unit wavelength")
plt.plot(lbda, b_val)
plt.plot(lbda, planck(lbda, *popt))
plt.show()
fle.close()
```
